```python
import numpy as np
t = np.arange(0, 5, 0.2)

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^') #three plots in one command: linear, square, cubic
# format for each is x-axis, y-axis, style

plt.legend(["line 1", "line 2", "line 3"])
```

```python
# Create a dataframe to use

import pandas as pd
df = pd.DataFrame(np.random.random((100,2)), columns = ['x', 'y'])
df.head()
```

```python
# Scatter plot
plt.scatter(df.x, df.y)
```

```python
# define a pattern: x > 0.2, y > 0.4
# (labelling data because we don't have data labels)

c = (df.x > 0.2) & (df.y > 0.4) # column with True/False printed based on these conditions
colors = np.where(c, 'r', 'g') # red within c, green outside c
plt.scatter(df.x, df.y, color = colors)
```

```python
# using subplots - will subplot from the plot before
# plt.subplot(# rows, # columns, subplot/position #) 

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t = np.arange(0, 5, 0.1)
plt.figure()
plt.subplot(2, 2, 1)
plt.plot(t, f(t), 'b-v')
plt.title("This will only title this one")
plt.subplot(2, 2, 4)
plt.plot(t, np.cos(2*np.pi*t), 'r--')
plt.suptitle('This titles all my waves', fontsize = 17) #titles entire thing - can have above all the lines if we want
# positions 2 and 3 are empty because we didn't add further subplots there
```

```python
# Bar charts

x, y = [1, 3, 5, 7, 9] , [7, 6, 1.5, 4, 6]
w, z = [2, 4, 6, 8, 10] , [8, 6, 2, 5, 8]

plt.bar(x, y, color = 'r', label = "Red")
plt.bar(w, z, color = 'k', label = "Black")
plt.legend()
plt.title("My Bar Graphs")
```
